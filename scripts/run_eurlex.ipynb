## Notebook usage example

This is an example of a notebook usage. You can write your code to test the library here

```python
import sys

from bulletin.eurlex.api.client import EurlexBulletinClient


client = EurlexBulletinClient()

date = "2025-01-01"
date_end = "2025-12-31"
title_contains = None # "research", "euronest"
language = "ENG"
category_type = "ANNOUNC" # Example: ["ANNOUNC", "RES", "NOTICE"]
institution_type = "COMP"  # Example: ["CONSIL", "COM", "MOVE", "COMP"]

try:
    acts = client.get_acts(date=date, language=language, date_end=date_end, title_contains=title_contains, category_type=category_type, institution_type=institution_type)
except Exception as exc:
    print(f"Error while fetching acts: {exc}", file=sys.stderr)

print("-" * 60)

for act in acts:
    print(act)
    print("-" * 60)

print(f"Total acts: {len(acts)}")
print("Done.")

```
