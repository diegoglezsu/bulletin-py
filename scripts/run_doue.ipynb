## Notebook usage example

This is an example of a notebook usage. You can write your code to test the library here

```python
import sys

from bulletin.doue.api.client import DoueBulletinClient


client = DoueBulletinClient()

date = "2026-01-01"
date_end = "2026-03-31"
#title_contains = "euronest"
language = "SPA"
category_type = "ANNOUNC"

try:
    acts = client.get_acts(date=date, language=language, date_end=date_end, title_contains=None, category_type=category_type)
except Exception as exc:
    print(f"Error while fetching acts: {exc}", file=sys.stderr)

print("-" * 60)

for act in acts:
    print(act)
    print("-" * 60)

print(f"Total acts: {len(acts)}")
print("Done.")

```

```python
# Add acts to a dataframe
csv = client.get_acts_csv(date=date, language=language, date_end=date_end, title_contains=None, category_type=category_type)
print("CSV Output:")
print(csv)
# You can use pandas to read the CSV output into a DataFrame
import pandas as pd
import io

df = pd.read_csv(io.StringIO(csv))
print(df.head())
```
